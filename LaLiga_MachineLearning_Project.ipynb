#Scrape Data for Multiple Teams for Multiple Years

```python
years = list(range(2023, 2020, -1))
```

```python
years
```

```python
all_matches = []
```

```python
standings_url = "https://fbref.com/en/comps/12/2023-2024/2023-2024-La-Liga-Stats"
```

```python
import time
for year in years:
  data = requests.get(standings_url)
  soup = BeautifulSoup(data.text)
  standings_table = soup.select('table.stats_table')[0]

  links = [l.get('href') for l in standings_table.find_all('a')]
  links = [l for l in links if '/squads/' in l]
  team_urls = [f"https://fbref.com{l}" for l in links]

  for team_url in team_urls:
    team_name = team_url.split('/')[-1].replace('-Stats', '').replace('-', ' ')
    data = requests.get(team_url)
    matches = pd.read_html(data.text, match="Scores & Fixtures")[0]
    soup = BeautifulSoup(data.text)
    links = [l.get("href") for l in soup.find_all('a')]
    links = [l for l in links if l and 'all_comps/shooting' in l]
    data  = requests.get(f"https://fbref.com{links[0]}")
    shooting = pd.read_html(data.text, match = "Shooting")[0]
    time.sleep(5)
    shooting.columns = shooting.columns.droplevel()
    try:
      team_data = matches.merge(shooting[["Date", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]], on="Date")
    except ValueError:
      continue
    team_data = team_data[team_data['Comp'] == "La Liga"]

    team_data['Season'] = year
    team_data["Team"] = team_name
    all_matches.append(team_data)
    time.sleep(15)
```

#Creating pedictors for machine learning

```python
matches['venue_code'] = matches['venue'].astype('category').cat.codes #taking into account the venue for home field advantage
```

```python
matches['opp_code'] = matches['opponent'].astype('category').cat.codes #creating codes for each specific opponent
```

```python
matches['hour'] = matches['time'].str.replace(":.+", "", regex=True).astype('int') #creating the hour column into integers to let the machine learning use this
```

```python
matches['day_code'] = matches['date'].dt.dayofweek #creating day of the week column
```

##Training Machine Learning Model

```python
matches['target'] = (matches['result'] == 'W').astype('int')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 50, min_samples_split = 10, random_state = 1)
```

```python
train = matches[matches['date'] < '2024-01-01'] #train variable is set to be any matches before 2023 so it can have all the data before so we can predict matches in the future
```

```python
test = matches[matches['date'] > '2024-01-01'] #test matches are matches in the future so it can predict matches after our sample space from our web scraping data
```

```python
predictors = ['venue_code', "opp_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train['target']) #.fit method will train a rf model with the predictors trying to predict the target
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test['target'], preds)
```

#Now we improve precision with rolling averages (this will increase the precision on future matches based on our data)

```python
grouped_matches = matches.groupby('team')
```

```python
group = grouped_matches.get_group('Real Madrid').sort_values('date')
```

```python
def rolling_averages(group, cols, new_cols):
  group = group.sort_values('date')
  rolling_stats = group[cols].rolling(3, closed = 'left').mean()
  group[new_cols] = rolling_stats
  group = group.dropna(subset = new_cols)
  return group
```

```python
cols = ['gf', 'ga', 'sh', 'sot', 'dist', 'fk', 'pk', 'pkatt']
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby('team').apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
def make_predictions(data, predictors):
  train = data[data['date'] < '2024-01-01']
  test = data[data['date'] > '2024-01-01']
  rf.fit(train[predictors], train['target'])
  preds = rf.predict(test[predictors])
  combined = pd.DataFrame(dict(actual=test['target'], predicted=preds), index=test.index)
  error = precision_score(test['target'], preds)
  return combined, error
```

```python
combined, error = make_predictions(matches_rolling, predictors + new_cols)
```
